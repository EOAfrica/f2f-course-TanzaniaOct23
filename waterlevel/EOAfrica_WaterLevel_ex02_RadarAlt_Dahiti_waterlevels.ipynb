## Step 1: Loading the necessary modules

```python
import xarray as xr
import pandas as pd
import matplotlib.pyplot as plt
```

## Step2: Get a timeseries of waterlevels for Lake Tana
- From the Database for Hydrological Time Series of Inland Waters (DAHITI)
- Link: https://dahiti.dgfi.tum.de/en/110/water-level/
- You need an account, and you can choose between different data formats.
- Or you simply use the .nc-file in the shared folder :)

```python
path = '../data/'
fn = '110_water_level_altimetry.nc'

tana = xr.open_dataset(path + fn)
```

```python
tana_time = pd.to_datetime(tana.date.values)
fig, ax = plt.subplots(figsize=(15,5))
ax.plot(tana_time, tana.water_level, '-o', markersize=3)
ax.set_ylabel('Water level in [m]')
ax.grid()
```
